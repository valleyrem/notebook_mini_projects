# **Comparison of Teaching Methods: Montessori vs. Traditional Approach**



---



> The study aimed to compare the effectiveness of teaching methods across various metrics, such as language and mathematical skills, memory, creativity, and well-being, between children educated using the Montessori system and traditional methods. Data were collected and analyzed for both preschool and school-age groups. The results reveal significant differences between the teaching methods in various domains, confirming the effectiveness of the Montessori method in several critical aspects of education.

*Total Participants*: 201 children (96 girls and 105 boys)

*Age Range*: 4.37 to 13.40 years old

*School Systems*: Montessori (n=99) and Traditional (n=102)

*Distribution*: Divided into kindergarten and elementary age groups

*Experimental Design*: Controlled observational study

*Comparison*: Montessori vs. Traditional schoolchildren

*Measures Used*: Various standardized tasks and questionnaires


# **Findings**:


*   Montessori children outperformed traditional schoolchildren in
language, math, well-being, working memory, convergent, and divergent creativity tasks at both kindergarten and elementary age levels.

* *Creativity skills* significantly mediated the relationship between school system (Montessori vs. Traditional) and academic outcomes, suggesting a partial mediation effect.

* Montessori education fosters a *more balanced global development*, promoting academic success beyond traditional measures of executive functions.

* Emphasizes the importance of nurturing creative execution abilities alongside traditional academic skills.

* Overall, the study provides insights into the scholastic benefits of Montessori education, highlighting the role of creativity skills in academic success and advocating for a more holistic approach to education.

#**Research data**

```python
from urllib.request import urlretrieve
import pandas as pd
data = pd.read_csv('csv-files/Montessori vs Traditional.csv')
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
selected_metrics = ['Language Skills', 'Math Skills', 'Memory', 'Creativity']
selected_data = data[data['Metric'].isin(selected_metrics)]

montessori_means = selected_data.loc[:, 'Montessori (Mean)']
traditional_means = selected_data.loc[:, 'Traditional (Mean)']

offset = 0.35
x = range(len(selected_metrics))
x2 = [i + offset for i in x]

plt.figure(figsize=(8, 5))
bars1 = plt.bar(x, montessori_means, width=0.35, color='lightgreen', label='Montessori')
bars2 = plt.bar(x2, traditional_means, width=0.35, color='red', alpha=0.5, label='Traditional')

for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

plt.xlabel('Metrics')
plt.ylabel('Mean Scores')
plt.title('Comparison of Montessori and Traditional Mean Scores')
plt.xticks(x, selected_metrics, rotation=30)

plt.legend(loc='upper center', bbox_to_anchor=(0.1, -0.3), fancybox=True, shadow=True, ncol=2)

plt.tight_layout()
plt.show()
```

```python
selected_metrics = ['Language Skills Preschool Age', 'Language Skills School Age',
                    'Math Skills Preschool Age', 'Math Skills School Age',
                    'Memory Preschool Age', 'Memory School Age',
                    'Creativity Preschool Age', 'Creativity School Age']

selected_data = data[data['Metric'].isin(selected_metrics)]
df = pd.DataFrame({
    'Metric': selected_data['Metric'],
    'Montessori': selected_data['Montessori (Mean)'],
    'Traditional': selected_data['Traditional (Mean)']
})

preschool_data = df[df['Metric'].str.contains('Preschool Age')]
school_data = df[df['Metric'].str.contains('School Age')]

plt.figure(figsize=(9, 5))

plt.subplot(1, 2, 1)
for idx, row in preschool_data.iterrows():
    plt.plot(['Montessori', 'Traditional'], [row['Montessori'], row['Traditional']], marker='o', label=row['Metric'])

plt.title('Preschool Age')
plt.xlabel('Category')
plt.ylabel('Mean Score')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
for idx, row in school_data.iterrows():
    plt.plot(['Montessori', 'Traditional'], [row['Montessori'], row['Traditional']], marker='o', label=row['Metric'])

plt.title('School Age')
plt.xlabel('Category')
plt.ylabel('Mean Score')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
